# While Loop
* Tests the condition as long as condition is True
* break will break out of the loop
* continue will bail out of the increment

````python
# Run the while at least 1 time
i = 5
while True: # This is an infinite loop
    print(i)
    if i >= 5:
        break
````

````python
min_length = 2
name = input("Please enter your name: ")

# If name is printable and is alphanumeric
while not(len(name) >= min_length and name.isprintable() and name.isalpha()):
    name = input("Please enter your name: ")

print('Hello, {0}'.format(name))
````

````python
# Refactored
min_length = 2
while True:
    name = input("Please enter your name: ")
    # If name is printable and is alphanumeric
    if len(name) >= min_length and name.isprintable() and name.isalpha():
        break
print('Hello, {0}'.format(name))
````

````python
# Continue allows for the loop to iterate back
# skip everything after the continue and go back to the loop
a = 0
while a < 10:
    a +=1
    if a%2 == 0 :
        continue
    print(a)
````

# Else clause for while executes the while loop and does not encounter a break


````python
# Else can be used with the While loop
# will execute if the while did NOT encounter break
l = [1, 2, 3]
val = 10
# Flag set to false
found = False
idx = 0
while idx < len(l):
    if l[idx] == val:
        found = True
        break
    idx += 1

if not found:
    l.append(val)
````

````python
# Without flag
l = [1, 2, 3, 10]
val = 10
idx = 0
while idx <len(l):
    if l[idx] == val:
        break
    idx += 1
else: # While loop terminated normally. There was no value that was found
    print('Value was in the list already')
    l.append(val)

print(l)
    
````
